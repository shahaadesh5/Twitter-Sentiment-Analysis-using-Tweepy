```python
import csv

analysis=[] #list for storing analysis result
score_values=[] #list for storing sentiment score

#separating non ASCII characters - https://github.com/stepthom/lexicon-sentiment-analysis/blob/master/doAnalysis.py

def strip_non_ascii(string):
    ''' Returns the string without non ASCII characters''' 
    stripped = (c for c in string if 0 < ord(c) < 127)
    return ''.join(stripped)

#function for removing hastags from the tweets

def remove_hash_from_string(string):
    without_hash=string.replace('#','')
    return without_hash

            
with open('tweets_file.csv',encoding='utf-8') as csvfile:
    readCSV = csv.reader(csvfile)
    csv_text = []
    
    #we need to clean the data based on ascii characters
    
    for row in readCSV:
        tweet=dict()
        tweet['text']=row[1]
        tweet['clean']=tweet['text']
        tweet = strip_non_ascii(tweet['clean'])
        tweet_hash=remove_hash_from_string(tweet)
        tweet_normalise=tweet_hash.lower()
        csv_text.append(tweet_hash)


lexicon = dict()  

#importing positive lexican file

with open('positive-words.csv',) as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    for row in readCSV:
        lexicon[row[0]]=[row[1]]
        
lexicon_neg = dict()

#importing negative lexican file

with open('negative-words.csv',) as csvfile:
    readCSV = csv.reader(csvfile, delimiter=',')
    for row in readCSV:
        lexicon_neg[row[0]]=[row[1]]
               
for tweet in csv_text:
   score = 0

    #comparing each tweet word with lexicans in respective files and updating score accordingly
   for word in tweet.split():
       if word in lexicon:
          
           score= score+ 1
           
       if word in lexicon_neg:
             
           score=score - 1
                   
   #appending score to score values list
   score_values.append(score)
   
   if(score >0):
            analysis.append("Positive")
   elif(score <0):
            analysis.append("Negative")
   else:
            analysis.append("Neutral")
        
        #transferring all saved data to sentiment csv file with the results
with open ('sentiment_file1.csv', 'w' , encoding = 'utf-8') as outfile:
    writer = csv.writer(outfile)
    writer.writerow(['Tweets','Sentiment','Sentiment Score'])
    for i in range(0,len(csv_text)):
        writer.writerow([csv_text[i],analysis[i],score_values[i]])
```
