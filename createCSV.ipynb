```python
import tweepy
import csv

consumer_key = ''
consumer_secret = ''
access_token = ''
access_token_secret = ''

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth,wait_on_rate_limit=True)
with open ('tweets_file.csv', 'w') as outfile:
    csvWriter = csv.writer(outfile)
    #tweepy cursor provides fetching of tweets with a particular keyword and number of tweets to fetch
    #keyword is realDonaldTrump and count indicates number of tweets to fetch
    for tweet in tweepy.Cursor(api.search,q="#realDonaldTrump",count=1000,
    lang="en").items(1000):
        #export the csv file with tweets and date of each tweet
        csvWriter.writerow([tweet.created_at, tweet.text.encode('utf-8')])
```
